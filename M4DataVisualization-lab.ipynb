```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
Q = """
SELECT ConvertedComp FROM master
"""
df = pd.read_sql_query(Q,conn)
df['ConvertedComp'].plot(kind='hist')
plt.title('Histogram of Compensation among Developers')
plt.xlabel('Compensation')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
Q = """
SELECT Age FROM master
"""
df = pd.read_sql_query(Q,conn)
df['Age'].plot(kind='box')
plt.title('Age Distribution of Developers')
plt.ylabel('Frequency')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
Q = """
SELECT * FROM master
"""
df = pd.read_sql_query(Q,conn)
df.plot(kind='scatter', x='Age', y='WorkWeekHrs')
plt.title('Correlation between Age and Weekly Working Hours')
plt.xlabel('Age')
plt.ylabel('Weekly Working Hours')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import seaborn as sns
Q = """
SELECT * FROM master
"""
df = pd.read_sql_query(Q,conn)
sns.scatterplot(data=df, x="WorkWeekHrs", y="CodeRevHrs", size="Age", sizes=(10, 1000), color='green')
plt.title('Weekly Code Review Hours vs Working Hours per Week')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import plotly.express as px
Q = """
SELECT COUNT(Respondent) AS C, DatabaseDesireNextYear FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY C DESC
"""
df = pd.read_sql_query(Q,conn)
df = df.head()
fig = px.pie(df, values='C', names='DatabaseDesireNextYear', title='Top 5 Databases Respondents Wish to Learn Next Year')
fig.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
Q = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35 
"""
df = pd.read_sql_query(Q,conn)
median_workweekhrs = df['WorkWeekHrs'].median()
median_coderevhrs = df['CodeRevHrs'].median()

list_name = [[median_workweekhrs, median_coderevhrs]]
df1 = pd.DataFrame (list_name, columns=['median_workweekhrs','median_coderevhrs'])
df1.plot(kind='bar', stacked=True)
plt.title('Median WorkWeekHrs and CodeRevHrs for the Age Group 30 to 35')
plt.xlabel('Age Group 30 to 35')
plt.ylabel('Median WorkWeekHrs and CodeRevHrs')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
Q = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(Q,conn)
line_data = df.groupby('Age')['ConvertedComp'].median()
line_data.plot(kind='line', figsize=(8,3))
plt.title('The Median ConvertedComp for All Ages from 45 to 60')
plt.ylabel('The Median ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
Q = """
SELECT COUNT(MainBranch) as C, MainBranch FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(Q,conn)
fig = px.bar(df, x='C', y='MainBranch', title='Professioncal Developers vs Individuals Who Simply Write Code ')
fig.show()
```
